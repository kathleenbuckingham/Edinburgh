```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#create dataframe
emissions_df = pd.read_csv("Resources/co2_emission.csv")
emissions_df.head()
```

```python
#understand dataframe
type(emissions_df)
emissions_df.shape
emissions_df.columns
pd.unique(emissions_df['Entity'])
emissions_df['Code'].describe()
```

```python
#Group data
grouped_data = emissions_df.groupby('Year')
grouped_data.describe()
emissions_df['Year'].min()
emissions_df['Year'].max()
```

```python
#understand emissions
emissions_df['Annual CO₂ emissions (tonnes )'].mean()
emissions_df['Annual CO₂ emissions (tonnes )'].std()
emissions_df['Annual CO₂ emissions (tonnes )'].count()
```

```python
#plot all countries 

for country in emissions_df['Entity'].unique():
    emissions_df[emissions_df['Entity']==country].sort_values('Year').plot.line(x='Year',
                                                           y='Annual CO₂ emissions (tonnes )',
                                                          title=country)
```

```python
country_year = emissions_df[emissions_df['Entity']=='Afganistan'].sort_values('Year')
co2 = emissions_df['Annual CO₂ emissions (tonnes )']

plt.plot(co2, country_year)
plt.xlabel('Year')
plt.ylabel('Annual CO2 emissions (tonnes')
plt.show()
```
